```python
# Bundles 2024 general results for display in 2024 MTFP election guide and Capitol Tracker
import pandas as pd
```

```python
xl = pd.ExcelFile('./sos-data/2024/general/2024-12-11-statewide-results.xlsx')
sheets = xl.sheet_names

sheets = {
    '01601PRESIDENT & VICE PRESIDEN': 'President',
    '01602UNITED STATES SENATOR': 'U.S. Senate',
    '01603UNITED STATES REPRESENTAT': 'U.S. House 1',
    '01604UNITED STATES REPRESENTAT': 'U.S. House 2',
    '01605GOVERNOR & LT. GOVERNOR': 'Governor',
    '01606SECRETARY OF STATE': 'Secretary of State',
    '01607ATTORNEY GENERAL': 'Attorney General',
    '01608STATE AUDITOR': 'State Auditor',
    '01609STATE SUPERINTENDENT OF P': 'Superintendent of Public Instruction',
    '01610PUBLIC SERVICE COMMISSION': 'PSC 2',
    '01611PUBLIC SERVICE COMMISSION': 'PSC 3',
    '01612PUBLIC SERVICE COMMISSION': 'PSC 4',
    '01613CLERK OF THE SUPREME COUR': 'Clerk of the Supreme Court',
    '01614SUPREME COURT CHIEF JUSTI': 'Supeme Court Chief Justice',
    '01615SUPREME COURT JUSTICE #3': 'Supreme Court Justice',
}


output = []
for sheet in sheets.keys():
    dfi = xl.parse(sheet)
    reporting = dfi.iloc[:,0][1]
    time = dfi.iloc[:,0][2]
    # race = dfi.iloc[:,0][5]
    race = sheets[sheet]
    print(race)
    # district = dfi.iloc[:,0][5].replace('STATE SENATOR','S').replace('STATE REPRESENTATIVE','H').replace(' DISTRICT','D').replace(' ','-')

    data = dfi[5:].drop(dfi.columns[0], axis='columns').transpose()
    data.columns = data.iloc[0]
    data.drop(data.index[0], inplace=True)
    data.reset_index(inplace=True)
    output.append(list(data.apply(lambda row: ({
        'race': race,
        'party': row['County'].split('\n')[1],
        'candidate': row['County'].split('\n')[0],
        'votes': row['TOTALS'],
    }), axis=1)))
    
output
```

```python
tall = pd.DataFrame([item for race in output for item in race])
tall
```

```python
# Output for Capitol Tracker district input
tall.to_csv('./outputs/2024/general/2024-general-statewide-vote-counts.csv')
```

```python
def summarize(row):
    total_votes = row['votes'].sum()
    gop_votes = row[row['party'] == 'Republican']['votes'].sum()
    dem_votes = row[row['party'] == 'Democrat']['votes'].sum()
    return pd.Series({
        'total_general_votes': total_votes,
        'fraction_general_votes_to_gop': gop_votes / total_votes,
        'gop_votes': gop_votes,
        'gop_candidate': ', '.join(row[row['party'] == 'Republican']['candidate']),
        'dem_votes': dem_votes,
        'dem_candidate': ', '.join(row[row['party'] == 'Democrat']['candidate']),
        
    }, index=['total_general_votes','fraction_general_votes_to_gop','gop_votes','gop_candidate','dem_votes','dem_candidate'])

summary = tall.groupby(['race']).apply(summarize).sort_values('fraction_general_votes_to_gop', ascending=False)
```

```python
# No Daves data here, unlike lege races
summary.to_csv('./outputs/2024/general/2024-statewide-general-by-party-vote.csv', index=False)
```
